```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix,classification_report
```

```python
df=pd.read_csv('/content/heart_disease_uci.csv')
df
```

```python
#  Convert 'num' into binary target: 0 (no disease), 1 (disease)
df['target']=df['num'].apply(lambda x:1 if x>0 else 0)
df
```

```python
# 3. Drop 'num' since we now have 'target'
df.drop('num',axis=1,inplace=True)
df
```

```python
# 4. Handle categorical values if present (e.g., 'sex', 'cp', etc.)
#    Convert non-numeric data to numbers (if any)
df_encoder=pd.get_dummies(df,drop_first=True)
df_encoder
```

```python
# 5. Handle missing values using SimpleImputer
imputer=SimpleImputer(strategy='mean')
df_imputed=pd.DataFrame(imputer.fit_transform(df_encoder),columns=df_encoder.columns)
df_imputed
```

```python
# 6. Split into features and target
x=df_imputed.drop('target',axis=1)
y=df_imputed['target']
```

```python
# 7. Split into training and testing sets
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#8. Standardize the features
Scaler=StandardScaler()
x_train=Scaler.fit_transform(x_train)
x_test=Scaler.transform(x_test)
```

```python
# 9. Train Logistic Regression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
# 10. Predict and evaluate
y_pred=model.predict(x_test)
```

```python
print(" Accuracy:", accuracy_score(y_test, y_pred))
```

```python
print("\n Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
print("\n Classification Report:\n", classification_report(y_test, y_pred))
```
